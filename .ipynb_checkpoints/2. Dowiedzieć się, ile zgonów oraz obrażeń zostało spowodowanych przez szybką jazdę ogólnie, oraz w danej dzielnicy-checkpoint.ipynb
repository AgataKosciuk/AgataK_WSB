# 2. Dowiedzieć się, ile zgonów oraz obrażeń zostało spowodowanych przez szybką jazdę ogólnie, oraz w danej dzielnicy

# A

```python
import pandas as pd
```

```python
motor_data = pd.read_csv("/Users/agata/Desktop/nypd-motor-vehicle-collisions.csv", low_memory=False, dtype={'third_column': 'str'})

motor_data
```

```python
all_borough_df1 = motor_data['NUMBER OF PERSONS INJURED']
all_borough_df2 = motor_data['NUMBER OF PERSONS KILLED']
all_borough_df3 = motor_data['NUMBER OF PEDESTRIANS INJURED']
all_borough_df4 = motor_data['NUMBER OF PEDESTRIANS KILLED']
all_borough_df5 = motor_data['NUMBER OF CYCLIST INJURED']
all_borough_df6 = motor_data['NUMBER OF CYCLIST KILLED']
all_borough_df7 = motor_data['NUMBER OF MOTORIST INJURED']
all_borough_df8 = motor_data['NUMBER OF MOTORIST KILLED']

display(all_borough_df1)
display(all_borough_df2)
display(all_borough_df3)
display(all_borough_df4)
display(all_borough_df5)
display(all_borough_df6)
display(all_borough_df7)
display(all_borough_df8)
```

```python
all_borough_df_concat = pd.concat([all_borough_df1, all_borough_df2, all_borough_df3, all_borough_df4, all_borough_df5, all_borough_df6, all_borough_df7, all_borough_df8])
all_borough_df_concat
```

# ODP: ilość zgonów i obrażeń ogólnie

```python
all_borough_df_concat.sum()
```

# B

```python
contain_queens = pd.DataFrame(motor_data, columns = ['BOROUGH', 'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED',
       'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED',
       'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED',
       'NUMBER OF MOTORIST KILLED'])
print (contain_queens)
```

```python
contain_queens2 = contain_queens[contain_queens['BOROUGH'].str.contains('QUEENS', na=False)]
print (contain_queens2)
```

```python
queens_borough_df_concat = pd.concat([contain_queens2])
queens_borough_df_concat
```

# ODP: ilość zgonów oraz obrażeń spowodowanych w wybranej dzielnicy - QUEENS

```python
queens_borough_df_concat.sum()
```
