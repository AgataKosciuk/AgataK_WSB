## Wybrana dzielnica QUEENS

```python
import pandas as pd
```

```python
motor_data = pd.read_csv("/Users/agata/Desktop/nypd-motor-vehicle-collisions.csv", low_memory=False, dtype={'third_column': 'str'})

motor_data
```

```python
contain_queens1 = pd.DataFrame(motor_data, columns = ['BOROUGH', 'CONTRIBUTING FACTOR VEHICLE 1',
       'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3',
       'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5'])
print (contain_queens1)
```

```python
contain_queens1= contain_queens1[contain_queens1['BOROUGH'].str.contains('QUEENS', na=False)]
print (contain_queens1)
```

```python
contain_queens1 = contain_queens1.dropna(how='all')
contain_queens1
```

```python
contain_queens1 = contain_queens1.dropna(how='any', axis=0)
contain_queens1
```

```python
contain_queens1.value_counts().nlargest(30)
contain_queens1
```
